```python
import numpy as np
```

```python
f = lambda x: np.sin(x**2)
```

```python
def _TR_(a,b,n,f):

    h = (b-a)/n
    TR = 0

    x = np.zeros((n+1,1))
    y = np.zeros((n+1,1))

    for i in range(1,n+1):
        x[i] = a+(i*h)
        y[i] = (f(x[i]))

    for k in range(1,n+1):
        TR += 0.5*h*(y[k-1] + y[k])

    return TR
```

```python
approx = [(N,_TR_(0,1,N,f)) for N in range(10, 1000) if N % 10 == 0]
```
